```python
categorical_miss_val = round(100*(data.select_dtypes(include=object).isna().sum().\
sort_values(ascending=False)/len(data.index)),2)\
.to_frame().rename(columns={0:'Null values percentage'})[:10]
categorical_miss_val
```

So we have null values in 3 columns
 - Poster
 - imdbVotes
 - imdbRating

```python
data['imdbVotes'] = data['imdbVotes'].str.replace(r'\W',"")
data['imdbRating'] = data['imdbRating'].str.replace(r'\W',"")
data['Year'] = data['Year'].str.replace(r'\W',"")
```

```python
#Filling the Null Values
data['imdbVotes'] = data['imdbVotes'].fillna(0)
data['imdbRating'] = data['imdbRating'].fillna(0)
data['Poster'] = data['Poster'].fillna('None')
```

```python
data["imdbRating"] = data["imdbRating"].replace('',0)
data["imdbVotes"] = data["imdbVotes"].replace('',0)
```

```python
data['imdbVotes'] = data['imdbVotes'].astype(int)
data['imdbRating'] = data['imdbRating'].astype(int)
data['Year'] = data['Year'].astype(int)
```
